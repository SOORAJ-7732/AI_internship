```python
from sklearn.cluster import KMeans
```

```python
K_range=range(1,15)
AK=[]
```

```python
for k in K_range:
    km = KMeans(n_clusters=k)
    km.fit(df[['QUANTITYORDERED','SALES']])
    AK.append(km.inertia_)  #Inertia_ is the sum of squared error for each cluster.
```

```python
AK
```

```python
plt.xlabel('Number of clusters (K)') 
plt.ylabel('Sum of squared error') 
plt.plot(K_range,AK) 
plt.title('My Shop') 
```
